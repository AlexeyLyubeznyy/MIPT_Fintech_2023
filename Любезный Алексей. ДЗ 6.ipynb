```python
ratings = pd.read_csv('/Users/alexeylyubeznyy/Documents/Магистратура/МФТИ/Advanced Python/6. Функции и работа с данными/ml-latest-small/ratings.csv')
ratings.info()
```

```python
movies = movies.rename(columns=str.lower)
ratings = ratings.rename(columns=str.lower)
```

```python
movies.head()
```

```python
ratings.head()
```

```python
ratings_pivot = ratings.pivot_table(index='movieid', values='rating', aggfunc='mean').sort_values(by='rating', ascending=False)
ratings_pivot
```

```python
def rating_f(row):
    if row[0]<=2:
        return 'Низкий рейтинг'
    
    elif row[0]<=4:
        return 'Средний рейтинг'
    
    elif row[0]>4:
        return 'Высокий рейтинг'
```

```python
ratings_pivot['class'] = ratings_pivot.apply(rating_f, axis=1)
ratings_pivot
```

```python
movies_ratings = movies.merge(ratings_pivot, on='movieid', how='inner')
movies_ratings
```

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск'],
    'Юг': ['ялта'] #в этом задании не приводил слова к исходной морфологической форме
}
```

```python
#keywords['keyword'].str.split()
```

```python
def region_f(row):
    for key,value in geo_data.items():
        for i in range(len(value)):
            if value[i] in row[0]:
                return key
            elif key == 'Юг':
                return 'undefined'
```

```python
keywords['region'] = keywords.apply(region_f, axis=1)
```
